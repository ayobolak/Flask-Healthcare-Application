```python
import pandas as pd

df = pd.read_csv('users.csv')
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='Age', y='Income')
plt.title('Income by Age')
plt.xlabel('Age')
plt.ylabel('Income (₦)')
plt.tight_layout()
plt.show()
```

```python
# Sort by income (ages with highest income)

top_income_by_age = df.sort_values(by='Income', ascending=False)

plt.figure(figsize=(8, 4))
sns.barplot(data=top_income_by_age, x='Age', y='Income', palette='viridis')

plt.title('Ages with the Highest Income')
plt.xlabel('Age')
plt.ylabel('Income (₦)')
plt.tight_layout()
plt.savefig('age_vs_income.png') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: style for better visuals
sns.set(style="whitegrid")

# Load the CSV file
df = pd.read_csv('users.csv')
df.head(30)
```

```python
# Gender distribution across spending categories

spend_cols = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
melted = df.melt(id_vars=['Gender'], value_vars=spend_cols,
                 var_name='Category', value_name='Amount')

plt.figure(figsize=(8, 4))
sns.barplot(data=melted, x='Category', y='Amount', hue='Gender', palette='Set2')

plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Expense Category')
plt.ylabel('Amount Spent (₦)')
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('gender_vs_spending.png')  # Save for PowerPoint
plt.show()
```
